```python
# Example DataFrame with numerical values
import pandas as pd

df = pd.DataFrame({
    'gender': [0, 1, 1, 0, 1]  # 0 = male, 1 = female
})

# Calculate mean of 'gender' column
mean_value = df['gender'].mean()

# Output the result
print("Mean value of 'gender':", mean_value)

```

```python
import pandas as pd

# Example DataFrame with numerical values
df = pd.DataFrame({
    'gender': [101,112,123,133]  # 0 = male, 1 = female
})

# Calculate median of 'gender' column
median_value = df['gender'].median()

# Output the result
print("Median value of 'gender':", median_value)
```
